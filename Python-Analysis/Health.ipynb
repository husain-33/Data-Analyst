```python
#importing all required libraries 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt

a = pd.read_csv('health.csv')
```

```python
#Returns Top 5 Rows of Table 
print(a.head())
```

```python
#Help us to null and data types of each rows 
print(a.info())
```

```python
# This described function is used to display all mathematics calculation of Column who contain numeric data
print(a.describe())
```

```python
print(a.isnull().sum())
```

```python
# Changing data type of Billing_Amount to integer 
a['Billing_Amount'] = a['Billing_Amount'].astype(int)
```

```python
print(a)
```

```python
# Count the patient with each Medical Conditions 
print(a['Medical_Condition'].value_counts())

```

```python
# Changing To Datetime data type 
a['Date_of_Admission'] = pd.to_datetime(a['Date_of_Admission'])
```

```python
#Counting Medical Condition in each year
a['year'] = a['Date_of_Admission'].dt.year
trend = a.groupby('year')['Medical_Condition'].value_counts()
print(trend)
```

```python
# Displaying All Columns of Table 
print(a.columns)
```

```python
# A pie chart which Display No. of male and female patient 
a['Gender'].value_counts().plot.pie(
    autopct='%1.1f%%',
    figsize=(6, 6),
    startangle=90,
    shadow=True,
    colors=['skyblue', 'lightcoral']  # optional
)
plt.title('Gender Distribution of Patients')
plt.ylabel('')
plt.show()
```

```python
#Average of Age 
a['Age'].mean()
```

```python
#Counting Patient with each age 
a['age_group'] = pd.cut(a['Age'], bins=[0,18,35,50,65,100], labels=['0-18','19-35','36-50','51-65','65+'])
a['age_group'].value_counts()
sns.countplot(x='age_group', data=a, color='Yellow')
```

```python
a['Date_of_Admission'] = pd.to_datetime(a['Date_of_Admission'],dayfirst=True)
a['Discharge_Date'] = pd.to_datetime(a['Discharge_Date'],dayfirst=True)
a['Stay_Duration'] = (a['Discharge_Date'] - a['Date_of_Admission']).dt.days
a['Stay_Duration'].describe()
```

```python
a.groupby('age_group')['Stay_Duration'].mean()
sns.countplot(x='age_group', data=a, color='Red')
```

```python
# 1. Get the top 3 most frequent conditions overall
top_conditions = a['Medical_Condition'].value_counts().head(10).index

# 2. Filter the DataFrame to only include those conditions
filtered_df = a[a['Medical_Condition'].isin(top_conditions)]

# 3. Group, unstack, and plot
filtered_df.groupby(['year', 'Medical_Condition']).size().unstack().plot(
    kind='line',
    figsize=(12, 6),
    marker='o' 
)

plt.title('Top 3 Medical Conditions Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.grid(True)
plt.legend(title='Condition')
plt.show()
```

```python
a.groupby(['year', 'Medical_Condition'])['Billing_Amount'].mean().unstack().plot(kind='line' ,figsize=(12, 6),
    marker='o')
plt.title(' Medical Conditions over Billing')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.grid(True)
plt.legend(title='Condition')
plt.show()
```
